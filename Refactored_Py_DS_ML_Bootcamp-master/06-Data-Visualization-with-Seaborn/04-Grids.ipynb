# Grids

Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
# print version
print("seaborn version: ", sns.__version__)
print("matplotlib version: ", matplotlib.__version__)

%autosave 600
```

```python
# for grid
# Add sns.set() before your code. After Seaborn v0.8, 
# we need to explicitly activate to escape Matplotlib defaults. 
# You may also try: sns.set_style("whitegrid")  or sns.set_style("dark").

sns.set()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# check out unique species
iris['species'].unique()
```

```python
# see pairplot first, use grid we can customize this plot
sns.pairplot(iris)
```

## PairGrid

Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

```python
# Just the Grid, like subplot for us
# pairplot does it automatically, pairgrid give us more control
sns.PairGrid(iris)
```

```python
# Then you map to the grid
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
#g.map_diag(plt.hist)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## pairplot

pairplot is a simpler version of PairGrid (you'll use quite often)

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

## Facet Grid

FacetGrid is the general way to create grids of plots based off of a feature:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Just the Grid
# g = sns.FacetGrid(data=tips, col="time", row="smoker")
g = sns.FacetGrid(tips, col="time", row="smoker")
```

```python
# there are 3 arguments at play here
g = sns.FacetGrid(tips, col="time",  row="smoker")
#g = g.map(plt.hist, "total_bill")
g = g.map(sns.distplot, "total_bill")
```

```python
# we know that sns.distplot only needs 1 argument
# lets see scatter plot which need 2 arguments
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Notice hwo the arguments come after plt.scatter call
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

## JointGrid

JointGrid is the general version for jointplot() type grids, for a quick example:

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```
